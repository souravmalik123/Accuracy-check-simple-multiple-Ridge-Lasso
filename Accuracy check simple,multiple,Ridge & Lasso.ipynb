```python
import pandas as pd
```

```python
data= pd.read_csv("mpg .txt")
```

```python
data.head(2)
```

```python
data.isnull().sum()
```

```python
x = data["weight"].values.reshape(-1,1)
print(x)
```

```python
y = data["mpg"].values.reshape(-1,1)
print(y)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
lr = LinearRegression()
```

```python
model_train = lr.fit(x_train,y_train)
print("The model training is completed")
```

```python
print("Accuracy of the training data")
model_train.score(x_train,y_train)  #### r2 value
```

```python
pred = model_train.predict(x_test)
pred
```

```python
##### check the accuracy of the testing 
data
from sklearn.metrics import r2_score
```

```python
 r2_score(y_test,pred)
```

## Ridge Regression

```python
rd = Ridge()
```

```python
model_train1 = rd.fit(x_train,y_train)
print("The model training is completed for Ridge")
```

```python
model_train1.score(x_train,y_train)
```

```python
pred1 = model_train1.predict(x_test)
r2_score(y_test,pred1)
```

```python
ls = Lasso()
```

```python
model_train2 = ls.fit(x_train,y_train)
print("The model training is completed for Lasso")
```

```python
model_train2.score(x_train,y_train)
```

```python
pred2 = model_train2.predict(x_test)
r2_score(y_test,pred2)
```
